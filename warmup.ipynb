```python
# Using numpy generate 1000 samples from standard normal
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.random.normal(0,1,1000)
```

```python
# Time series plot
plt.figure(figsize=(10, 6))
plt.plot(samples)
plt.title('Time Series Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, edgecolor='black') 
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
